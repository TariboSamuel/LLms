#Gemini API Generation Parameters

```python
# importing the libraries
import google.generativeai as genai
import os
from dotenv import load_dotenv, find_dotenv

# loading the API key and authenticating to Gemini
load_dotenv(find_dotenv(), override=True)
os.environ.get('GOOGLE_API_KEY')

# configuring the API key.
genai.configure(api_key=os.environ.get('GOOGLE_API_KEY'))

# selecting the Gemini PRO model
model = genai.GenerativeModel('gemini-pro')

# generating content for a prompt
prompt = 'write me a poem to a lovely, fair and beautiful damsel. not flirting'
response = model.generate_content(prompt)

# displaying the response
print(response.text)
```

```python
# importing the libraries
import google.generativeai as genai
import os
from dotenv import load_dotenv, find_dotenv

# loading the API key and authenticating to Gemini
load_dotenv(find_dotenv(), override=True)
os.environ.get('GOOGLE_API_KEY')

# configuring the API key.
genai.configure(api_key=os.environ.get('GOOGLE_API_KEY'))

# selecting the Gemini PRO model
model = genai.GenerativeModel('gemini-pro')

# generating content for a prompt
prompt = "write me a short note to a lovely, fair and beautiful girl, not flirting her name's victoria and she's really nice and sweet, i just wanna tell  how i feel about her and that she's amazing. i haven't really spent time physically with her but these days, i don't even have time for myself "
response = model.generate_content(prompt)

# displaying the response
print(response.text)
```

```python
# defining a GenerationConfig object
generation_config = genai.types.GenerationConfig() # with default options

# with different options
# generation_config = genai.teyps.GenerationConfig(
#     canditate_count=1,
#     stop_sequence=[','],
#     max_output_tokens=32000,
#     temperature=0.9,
#     top_p=1,
#     top_k=1
# )

# use the GeenerationConfig object in an API call with either:
# 1.
model = genai.GenerativeModel('gemini-pro', generation_config=generation_config)
# or 2.
response = model.generate_content(prompt, generation_config=generation_config)
```
